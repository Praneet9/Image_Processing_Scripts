## Lane Detection on an Image


### Importing Packages

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading the Image
Opencv uses image in BGR format i.e. Blue Green Red rather than standard way of RGB

Matplotlib uses the standard RGB format, so we need to convert to BGR to plot appropriate colours using Matplotlib

```python
image = cv2.imread('lane.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

### Function to find the region of interest in the image for us to work on

```python
def region_of_interest(img, vertices):
    mask = np.zeros_like(img)
    
    try:
        channels = img.shape[2]
        match_mask_color = (255,) * channels
    except IndexError as _:
        match_mask_color = 255

    cv2.fillPoly(mask, vertices, match_mask_color)
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
```

### Preprocessing
Converting to grayscale and detecting edges using Canny Algorithm to find the lines in the image

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

canny = cv2.Canny(gray, 100, 200)

plt.imshow(canny, cmap='gray')
```

### Discarding unnecessary part and keeping only the Region of interest

```python
region_of_interest_vertices = [
    (0, image.shape[0]),
    (image.shape[1] / 2, image.shape[0] / 2),
    (image.shape[1], image.shape[0]),
]

roi = region_of_interest(canny, np.array([region_of_interest_vertices], np.int32))
plt.imshow(roi, cmap='gray')
```

### Detecting lines in the image using Hough Transform Algorithm
The parameters used are manually set as per need. You can play around with them.

```python
lines = cv2.HoughLinesP(roi, rho=2, theta = np.pi / 45, threshold = 160, minLineLength = 80, maxLineGap=25)
```

### Function to Draw lane lines on the image

```python
def drawLines(img, lines, colour = (255, 0, 0), thickness = 3):
    if lines is None:
        return None
    
    img = np.copy(img)
    
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    
    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(line_img, (x1, y1), (x2, y2), colour, thickness)
            
    img = cv2.addWeighted(img, 0.8, line_img, 1.0, 0.0)
    
    return img
```

### Plotting the detected lines in the image

```python
final = drawLines(image, lines)

plt.imshow(final)
```

### Seperating into Two lines - Left and Right

```python
left_line_x = []
left_line_y = []
right_line_x = []
right_line_y = []
```

### Parsing through lines detected
Seperating lines with negative slope as left line and positive slope as right line

```python
for line in lines:
    x1, y1, x2, y2 = line[0]
    slope = (y2 - y1) / (x2 - x1)
    if abs(slope) < 0.5:
        continue
    if slope <= 0:
        left_line_x.extend([x1, x2])
        left_line_y.extend([y1, y2])
    else:
        right_line_x.extend([x1, x2])
        right_line_y.extend([y1, y2])
```

```python
min_y = int(image.shape[0] * (3 / 5))
max_y = image.shape[0]
```

```python
poly_left = np.poly1d(np.polyfit(
    left_line_y,
    left_line_x,
    deg = 1
))

left_x_start = int(poly_left(max_y))
left_x_end = int(poly_left(min_y))

poly_right = np.poly1d(np.polyfit(
    right_line_y,
    right_line_x,
    deg=1
))

right_x_start = int(poly_right(max_y))
right_y_end = int(poly_right(min_y))
```

### Plotting Lanes detected in the image

```python
line_image = drawLines(
    image,
    [[
        [left_x_start, max_y, left_x_end, min_y],
        [right_x_start, max_y, right_y_end, min_y]
    ]],
    thickness=5,
)

plt.imshow(line_image)
```



## Lane Detection on a video

### Function to get the Region on Interest

```python
def region_of_interest(img, vertices):
    mask = np.zeros_like(img)
    
    try:
        channels = img.shape[2]
        match_mask_color = (255,) * channels
    except IndexError as _:
        match_mask_color = 255

    cv2.fillPoly(mask, vertices, match_mask_color)
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
```

### Function to Draw lines on the image

```python
def drawLines(img, lines, colour = (0, 0, 255), thickness = 3):
    if lines is None:
        return None
    
    img = np.copy(img)
    
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    
    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(line_img, (x1, y1), (x2, y2), colour, thickness)
            
    img = cv2.addWeighted(img, 0.8, line_img, 1.0, 0.0)
    
    return img
```

### Engine Room for lane detection
All the processing is done in this function

```python
def detect_lanes(frame):
#     frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    canny = cv2.Canny(gray, 100, 200)
    
    height = frame.shape[0]
    width = frame.shape[1]
    
    region_of_interest_vertices = [
    (0, height),
    (width / 2, height / 2),
    (width, height)]

    roi = region_of_interest(canny, np.array([region_of_interest_vertices], np.int32))
    
    lines = cv2.HoughLinesP(roi, rho=6,
                            theta = np.pi / 45, 
                            threshold = 160, 
                            minLineLength = 80, 
                            maxLineGap=25)
    
    left_line_x = []
    left_line_y = []
    right_line_x = []
    right_line_y = []
    
    for line in lines:
        x1, y1, x2, y2 = line[0]
        slope = (y2 - y1) / (x2 - x1)
        if abs(slope) < 0.5:
            continue
        if slope <= 0:
            left_line_x.extend([x1, x2])
            left_line_y.extend([y1, y2])
        else:
            right_line_x.extend([x1, x2])
            right_line_y.extend([y1, y2])
            
    min_y = int(height * (3 / 5))
    max_y = height

    if not left_line_x or not left_line_y or not right_line_x or not right_line_y:
        return frame
    poly_left = np.poly1d(np.polyfit(
        left_line_y,
        left_line_x,
        deg = 1
    ))

    left_x_start = int(poly_left(max_y))
    left_x_end = int(poly_left(min_y))

    poly_right = np.poly1d(np.polyfit(
        right_line_y,
        right_line_x,
        deg=1
    ))

    right_x_start = int(poly_right(max_y))
    right_y_end = int(poly_right(min_y))
    
    line_image = drawLines(
        frame,
        [[
            [left_x_start, max_y, left_x_end, min_y],
            [right_x_start, max_y, right_y_end, min_y]
        ]],
        thickness=5,
    )
    
    return line_image
```

### Capturing video and processing frame by frame

```python
cap = cv2.VideoCapture('video.mp4')
 
# Check if video opened successfully
if (cap.isOpened()== False): 
    print("Error opening video stream or file")

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
out = cv2.VideoWriter('output.mp4',cv2.VideoWriter_fourcc(*'XVID'), 10, (frame_width,frame_height))

# Read until video is completed
while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:

        # Display the resulting frame
        lanes = detect_lanes(frame)
        cv2.imshow('Frame',lanes)

        # Write the frame into the file 'output.avi'
        out.write(lanes)
        
        # Press Q on keyboard to  exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    # Break the loop
    else: 
        break

# When everything done, release the video capture object
cap.release()
 
# Closes all the frames
cv2.destroyAllWindows()
```
